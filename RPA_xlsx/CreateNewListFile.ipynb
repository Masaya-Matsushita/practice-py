```python
#リファクタリング(関数切り出し)

def get_invoice_data(file_path: Path) -> (datetime.date, str, int):
    wb = load_workbook(file_path, data_only=True)
    ws = wb.worksheets[0]
    inv_datetime = ws['H4'].value
    #日付
    inv_date = datetime.date(
        inv_datetime.year,
        inv_datetime.month,
        inv_datetime.day
        )
    #　請求先
    inv_company = ws['A7'].value
    # 金額
    inv_amount = ws['B13'].value
    return inv_date, inv_company, inv_amount

def main():
    folder = Path(r'/Users/matsushitamasaya/study/practice-py/RPA_xlsx')
    
    df = pd.DataFrame()
    for file_path in folder.grpb('*.xlsx'):
        # 請求書データ抽出
        inv_date, inv_com, inv_amo = get_invoice_data(file_path)
        df = df.append({
            '日付': inv_date,
            '請求先': inv_com,
            '金額': inv_amo
        }, ignore_index=True)
    df.to_excel('2022.xlsx')


if __name__ == '__main__':
    main()
```
